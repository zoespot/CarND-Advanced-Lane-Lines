## Camera Calibration

```python
#get camera calibaration mtx, dist
objpoints = [] #3D points in real world 
imgpoints =[] #2D points in image plane
#prepare object points like (0,0,0),(1,0,0)...(8,5,0) 9x6
nx = 9 # the number of inside corners in x
ny = 6 # the number of inside corners in y
objp =np.zeros((ny*nx,3),np.float32)
objp[:,:2] = np.mgrid[0:nx,0:ny].T.reshape(-1,2) # x,y coordinates

for name in glob.glob('camera_cal/*'):
    img = cv2.imread(name)
    #convert image to grayscale
    gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    #Find chessboard corners
    ret, corners = cv2.findChessboardCorners(gray, (nx,ny), None)
    #If corneres  found 
    if ret == True:
        imgpoints.append(corners)
        objpoints.append(objp)

#calibrate camera 
ret, mtx, dist, rvecs, tvecs = cv2.calibrateCamera(objpoints, imgpoints, gray.shape[::-1], None, None)
```

## Thresholding
Use combinations of color and gradient thresholds to generate a binary image where the lane lines are clearly visible.

```python
def binary_threshold(img,mtx,dist):
    # undistort image with camera calibration matrix
    undistorted = cv2.undistort(img, mtx, dist, None, mtx)

    # Convert to HLS color space
    hls = cv2.cvtColor(undistorted, cv2.COLOR_RGB2HLS)
    h_channel = hls[:,:,0]
    l_channel = hls[:,:,1]
    s_channel = hls[:,:,2]
    
    # Convert to HSV color space
    hsv = cv2.cvtColor(undistorted, cv2.COLOR_RGB2HSV)
    ss_channel = hsv[:,:,1]
    vv_channel = hsv[:,:,2]

    # Grayscale image
    gray = cv2.cvtColor(undistorted, cv2.COLOR_RGB2GRAY)

   # Sobel x
    sobelx = cv2.Sobel(gray, cv2.CV_64F, 1, 0) # Take the derivative in x
    abs_sobelx = np.absolute(sobelx) # Absolute x derivative to accentuate lines away from horizontal
    scaled_sobelx = np.uint8(255*abs_sobelx/np.max(abs_sobelx))

    # Threshold x gradient 20-100
    thresh_min = 20
    thresh_max = 100
    sxbinary = np.zeros_like(scaled_sobelx)
    sxbinary[(scaled_sobelx >= thresh_min) & (scaled_sobelx <= thresh_max)] = 1
    
    # Sobel y
    sobely = cv2.Sobel(gray, cv2.CV_64F, 0, 1) # Take the derivative in x
    abs_sobely = np.absolute(sobely) # Absolute x derivative to accentuate lines away from horizontal
    scaled_sobely = np.uint8(255*abs_sobelx/np.max(abs_sobelx))

    # Threshold y gradient 20-100
    thresh_min = 10
    thresh_max = 100
    sybinary = np.zeros_like(scaled_sobely)
    sybinary[(scaled_sobely >= thresh_min) & (scaled_sobely <= thresh_max)] = 1   
    
    # Threshold color channel 
    s_thresh_min = 190
    s_thresh_max = 210
    s_binary = np.zeros_like(s_channel)
    s_binary[(s_channel >= s_thresh_min) & (s_channel <= s_thresh_max)] = 1
    
    # Threshold color channel 
    h_thresh_min = 20
    h_thresh_max = 30
    h_binary = np.zeros_like(h_channel)
    h_binary[(h_channel >= h_thresh_min) & (h_channel <= h_thresh_max)] = 1

    # Threshold color channel 
    l_thresh_min = 200
    l_thresh_max = 250
    l_binary = np.zeros_like(l_channel)
    l_binary[(l_channel >= l_thresh_min) & (l_channel <= l_thresh_max)] = 1
    
    # Threshold color channel 
    ss_thresh_min = 180
    ss_thresh_max = 200
    ss_binary = np.zeros_like(ss_channel)
    ss_binary[(ss_channel >= ss_thresh_min) & (ss_channel <= ss_thresh_max)] = 1
    
        # Threshold color channel 
    vv_thresh_min = 220
    vv_thresh_max = 255
    vv_binary = np.zeros_like(vv_channel)
    vv_binary[(vv_channel >= vv_thresh_min) & (vv_channel <= vv_thresh_max)] = 1

    # Combine the two binary thresholds
    combined_binary = np.zeros_like(sxbinary)
    combined_binary[(s_binary == 1)|(l_binary == 1) |(vv_binary==1)] = 1
    
    return combined_binary, sxbinary, sybinary, h_binary, l_binary, s_binary, ss_binary, vv_binary
#     return  combined_binary
```

## Perspective Transform

```python
def perspective_transform(img,src_pts,dst_pts):  

    M = cv2.getPerspectiveTransform(src_pts,dst_pts)
    Minv = cv2.getPerspectiveTransform(dst_pts,src_pts)
    img_size= (img.shape[1],img.shape[0])
    warped =cv2.warpPerspective(img, M, img_size, flags=cv2.INTER_LINEAR)
    
    return warped, M, Minv
```

## Line Class Definition

```python
# Define a class to receive the characteristics of each line detection
class Line():
    def __init__(self):
        # was the line detected in the last iteration?
        self.detected = False  
        # x values of the last n fits of the line
        self.recent_xfitted = [] 
        #average x values of the fitted line over the last n iterations
        self.bestx = None     
        #polynomial coefficients averaged over the last n iterations
        self.best_fit = None  
        #polynomial coefficients for the most recent fit
        self.current_fit = [] # [np.array([False])]  
        #radius of curvature of the line in some units
        self.radius_of_curvature = None 
        #distance in meters of vehicle center from the line
        self.line_base_pos = None 
        #difference in fit coefficients between last and new fits
        self.diffs = np.array([0,0,0], dtype='float') 
        #x values for detected line pixels
        self.allx = None  
        #y values for detected line pixels
        self.ally = None
        #whether the point is newly added 
        self.new = False
```

## Function to search lane pixels from scratch

```python
#Function to find lane pixels from warped image from scratch
def search_new_poly(binary_warped):

    # Take a histogram of the bottom half of the image, floor divide
    histogram = np.sum(binary_warped[binary_warped.shape[0]//2:,:], axis=0)
    # Create an output image to draw on and visualize the result
    out_img = np.dstack((binary_warped, binary_warped, binary_warped))
    # Find the peak of the left and right halves of the histogram
    # These will be the starting point for the left and right lines
    midpoint = np.int(histogram.shape[0]//2)
    leftx_base = np.argmax(histogram[:midpoint])
    rightx_base = np.argmax(histogram[midpoint:]) + midpoint

    # HYPERPARAMETERS
    # Choose the number of sliding windows
    nwindows = 9
    # Set the width of the windows +/- margin
    margin = 100
    # Set minimum number of pixels found to recenter window
    minpix = 50

    # Set height of windows - based on nwindows above and image shape
    window_height = np.int(binary_warped.shape[0]//nwindows)
    # Identify the x and y positions of all nonzero pixels in the image
    nonzero = binary_warped.nonzero()
    nonzeroy = np.array(nonzero[0])
    nonzerox = np.array(nonzero[1])
    # Current positions to be updated later for each window in nwindows
    leftx_current = leftx_base
    rightx_current = rightx_base

    # Create empty lists to receive left and right lane pixel indices
    left_lane_inds = []
    right_lane_inds = []

    # Step through the windows one by one
    for window in range(nwindows):
        # Identify window boundaries in x and y (and right and left)
        win_y_low = binary_warped.shape[0] - (window+1)*window_height
        win_y_high = binary_warped.shape[0] - window*window_height
        ### Find the four below boundaries of the window ###
        win_xleft_low = leftx_current - margin  # Update this
        win_xleft_high = leftx_current + margin  # Update this
        win_xright_low = rightx_current -margin  # Update this
        win_xright_high = rightx_current +margin # Update this
              
        ### Identify the nonzero pixels in x and y within the window ###
        good_left_inds = ((nonzeroy >= win_y_low) & (nonzeroy < win_y_high) & 
        (nonzerox >= win_xleft_low) &  (nonzerox < win_xleft_high)).nonzero()[0]
        good_right_inds = ((nonzeroy >= win_y_low) & (nonzeroy < win_y_high) & 
        (nonzerox >= win_xright_low) &  (nonzerox < win_xright_high)).nonzero()[0]
        #print(type(good_right_inds),type(good_right_inds[0]))
        # Append these indices to the lists
        left_lane_inds.append(good_left_inds)
        right_lane_inds.append(good_right_inds)
        
        ### If you found > minpix pixels, recenter next window ###
        ### (`right` or `leftx_current`) on their mean position ###
        if len(good_left_inds) > minpix:
            leftx_current = np.int(np.mean(nonzerox[good_left_inds]))
        if len(good_right_inds) > minpix:        
            rightx_current = np.int(np.mean(nonzerox[good_right_inds]))
        #print(len(left_lane_inds),np.concatenate(left_lane_inds).shape)
    # Concatenate the arrays of indices (previously was a list of lists of pixels)
    try:
        left_lane_inds = np.concatenate(left_lane_inds)
        right_lane_inds = np.concatenate(right_lane_inds)
    except ValueError:
        # Avoids an error if the above is not implemented fully
        pass

    # Extract left and right line pixel positions
    leftx = nonzerox[left_lane_inds]
    lefty = nonzeroy[left_lane_inds] 
    rightx = nonzerox[right_lane_inds]
    righty = nonzeroy[right_lane_inds]
    
    ### Fit a second order polynomial to each using `np.polyfit` ###
    left_fit = np.polyfit(lefty,leftx,2)
    right_fit = np.polyfit(righty,rightx,2)

    # Generate x and y values for plotting
    ploty = np.linspace(0, binary_warped.shape[0]-1, binary_warped.shape[0] )

    try:
        left_fitx = left_fit[0]*ploty**2 + left_fit[1]*ploty + left_fit[2]
        right_fitx = right_fit[0]*ploty**2 + right_fit[1]*ploty + right_fit[2]
    except TypeError:
        # Avoids an error if `left` and `right_fit` are still none or incorrect
        print('The function failed to fit a line!')
        left_fitx = 1*ploty**2 + 1*ploty
        right_fitx = 1*ploty**2 + 1*ploty
        
    return left_fit, right_fit, left_fitx, right_fitx, ploty
```

##  Function to search lane pixel based on existing fit

```python
#Skip the sliding windows step once you've found the lines
def search_around_poly(binary_warped, left_fit,right_fit):
    # HYPERPARAMETER
    # Choose the width of the margin around the previous polynomial to search
    # The quiz grader expects 100 here, but feel free to tune on your own!
    margin = 100

    # Grab activated pixels
    nonzero = binary_warped.nonzero()
    nonzeroy = np.array(nonzero[0])
    nonzerox = np.array(nonzero[1])
    
    ### TO-DO: Set the area of search based on activated x-values ###
    ### within the +/- margin of our polynomial function ###
    ### Hint: consider the window areas for the similarly named variables ###
    ### in the previous quiz, but change the windows to our new search area ###
    left_lane_inds =((nonzerox > (left_fit[0]*(nonzeroy**2) + left_fit[1]*nonzeroy + 
                    left_fit[2] - margin)) & (nonzerox < (left_fit[0]*(nonzeroy**2) + 
                    left_fit[1]*nonzeroy + left_fit[2] + margin)))
    right_lane_inds = ((nonzerox > (right_fit[0]*(nonzeroy**2) + right_fit[1]*nonzeroy + 
                    right_fit[2] - margin)) & (nonzerox < (right_fit[0]*(nonzeroy**2) + 
                    right_fit[1]*nonzeroy + right_fit[2] + margin)))
    
    # Again, extract left and right line pixel positions
    leftx = nonzerox[left_lane_inds]
    lefty = nonzeroy[left_lane_inds] 
    rightx = nonzerox[right_lane_inds]
    righty = nonzeroy[right_lane_inds]
   
    ### Fit a second order polynomial to each using `np.polyfit` ###
    left_fit = np.polyfit(lefty,leftx,2)
    right_fit = np.polyfit(righty,rightx,2)

    # Generate x and y values for plotting
    ploty = np.linspace(0, binary_warped.shape[0]-1, binary_warped.shape[0] )

    try:
        left_fitx = left_fit[0]*ploty**2 + left_fit[1]*ploty + left_fit[2]
        right_fitx = right_fit[0]*ploty**2 + right_fit[1]*ploty + right_fit[2]
    except TypeError:
        # Avoids an error if `left` and `right_fit` are still none or incorrect
        print('The function failed to fit a line!')
        left_fitx = 1*ploty**2 + 1*ploty
        right_fitx = 1*ploty**2 + 1*ploty
        
    return left_fit, right_fit, left_fitx, right_fitx, ploty
```

## Function to chooose which search algorithm to use

```python
#Choose which search algorithm to use based on Line class information 
def find_lane_pixels(binary_warped,Left_Lines, Right_Lines):
    if Left_Lines!=[]:
        if(Left_Lines[-1].best_fit != [] and Right_Lines[-1].best_fit !=[] and Left_Lines[-1].new ==True and Right_Lines[-1]==True):
            left_fit, right_fit, left_fitx, right_fitx, ploty = search_around_poly(binary_warped, Left_Lines[-1].best_fit,Right_Lines[-1].best_fit)
    else: 
        left_fit, right_fit, left_fitx, right_fitx, ploty = search_new_poly(binary_warped)
    return left_fit, right_fit, left_fitx, right_fitx, ploty
```

```python
#draw lanes with area and curvation and offset information
def draw_lines(left_fit, right_fit,left_fitx, right_fitx, ploty):
    
    left_line=Line()
    n =3 #past number of records   
    
    if left_fitx!=[]:
        left_line.detected =True
        if len(Left_Lines)==0:
            left_line.recent_xfitted = left_fitx
            left_line.bestx =  np.average(left_line.recent_xfitted)
            left_line.current_fit = left_fit
            left_line.best_fit = left_fit
            left_line.diffs =0
        elif len(Left_Lines) < n:
            left_line.recent_xfitted = np.append(Left_Lines[-1].recent_xfitted.reshape(-1,left_fitx.shape[0]), left_fitx.reshape(1,-1), axis=0) 
            left_line.bestx = np.average(left_line.recent_xfitted)
            left_line.current_fit = left_fit
            left_line.best_fit = np.average([Left_Lines[-1].current_fit, left_fit], axis=0)
#             left_line.diffs = np.divide((Left_Lines[-1].current_fit - left_line.best_fit),Left_Lines[-1].current_fit)
            left_line.diffs = np.sum(np.square(Left_Lines[-1].current_fit - left_line.best_fit)) 
        else:
            left_line.recent_xfitted = np.append(Left_Lines[-1].recent_xfitted[-n:,:], left_fitx.reshape(1,-1), axis=0) 
            left_line.bestx = np.average(left_line.recent_xfitted)
            left_line.current_fit = left_fit
            left_line.best_fit = np.average([Left_Lines[-1].current_fit, left_fit], axis=0)
#             left_line.diffs = np.divide((Left_Lines[-1].current_fit - left_line.best_fit),Left_Lines[-1].current_fit)       
            left_line.diffs = np.sum(np.square(Left_Lines[-1].current_fit - left_line.best_fit)) 
            
        left_line.allx = left_line.best_fit[0] *ploty*ploty + left_line.best_fit[1]*ploty +left_line.best_fit[2]
        left_line.ally = ploty

        left_fit_cr = np.polyfit(left_line.ally*ym_per_pix, left_line.allx*xm_per_pix, 2)
        left_y_eval = np.max(left_line.ally)*ym_per_pix
        left_line.radius_of_curvature = (1+(2*left_fit_cr[0]*left_y_eval+left_fit_cr[1])**2)**1.5/(2*np.absolute(left_fit_cr[0])) 
        
    right_line=Line()
    n =3 #past number of records
    if right_fitx!=[]:
        right_line.detected =True
        if len(Right_Lines)==0:
            right_line.recent_xfitted = right_fitx
            right_line.bestx =  np.average(right_line.recent_xfitted)
            right_line.current_fit = right_fit
            right_line.best_fit = right_fit
            right_line.diffs =0
        elif len(Right_Lines) < n:
            right_line.recent_xfitted = np.append(Right_Lines[-1].recent_xfitted.reshape(-1,right_fitx.shape[0]), right_fitx.reshape(1,-1), axis=0) 
            right_line.bestx = np.average(right_line.recent_xfitted)
            right_line.current_fit = right_fit
            right_line.best_fit = np.average([Right_Lines[-1].current_fit, right_fit], axis=0)
#             right_line.diffs = np.divide((Right_Lines[-1].current_fit - right_line.best_fit),Right_Lines[-1].current_fit)
            right_line.diffs = np.sum(np.square(Right_Lines[-1].current_fit - right_line.best_fit)) 
        else:
            right_line.recent_xfitted = np.append(Right_Lines[-1].recent_xfitted[-n:,:], right_fitx.reshape(1,-1), axis=0) 
            right_line.bestx = np.average(right_line.recent_xfitted)
            right_line.current_fit = right_fit
            right_line.best_fit = np.average([Right_Lines[-1].current_fit, right_fit], axis=0)
#             right_line.diffs = np.divide((Right_Lines[-1].current_fit - right_line.best_fit),Right_Lines[-1].current_fit)  
            right_line.diffs = np.sum(np.square(Right_Lines[-1].current_fit - right_line.best_fit))  
 
        right_line.allx = right_line.best_fit[0] *ploty*ploty + right_line.best_fit[1]*ploty +right_line.best_fit[2]
        right_line.ally = ploty
        right_fit_cr = np.polyfit(right_line.ally*ym_per_pix, right_line.allx*xm_per_pix, 2)
        right_y_eval = np.max(right_line.ally)*ym_per_pix
        right_line.radius_of_curvature = (1+(2*right_fit_cr[0]*right_y_eval+right_fit_cr[1])**2)**1.5/(2*np.absolute(right_fit_cr[0])) 
    
    #sanity check 
    if len(Left_Lines)>1:
        if left_line.diffs>2000:
            left_line.detected = False
        if right_line.diffs >2000:
            right_line.detected =False
        if ((right_line.bestx - left_line.bestx) > Xpixel*1.3 or (right_line.bestx - left_line.bestx) <Xpixel*0.7):
            left_line.detected = False
            right_line.detected =False
        if (np.divide(right_line.radius_of_curvature,left_line.radius_of_curvature)>5 or np.divide(right_line.radius_of_curvature,left_line.radius_of_curvature)<0.2):
            left_line.detected= False
            right_line.detected=False 
#     print('right_line.radius_of_curvature ', right_line.radius_of_curvature ,left_line.radius_of_curvature )  
#     print('radius ratio',np.divide(right_line.radius_of_curvature,left_line.radius_of_curvature))
#     print('bestxdiff',right_line.bestx - left_line.bestx)
#     print('right_line.diffs',right_line.diffs,left_line.diffs)
#     print(right_line.detected)
#     print(len(Right_Lines))
    if (left_line.detected ==True) and (right_line.detected ==True):
        left_line.line_base_pos = (0.5*binary_warped.shape[1] - (np.average(left_fitx)+ np.average(right_fitx))/2) *xm_per_pix
        right_line.line_base_pos = left_line.line_base_pos
        left_line.new= True
        right_line.new=True
        Right_Lines.append(right_line)        
        Left_Lines.append(left_line) 
        
    return Left_Lines, Right_Lines
```

## Function to draw the area and curvation/offset information 

```python
def draw_area(img, Left_Lines, Right_Lines):    
    
    undistorted = cv2.undistort(img, mtx, dist, None, mtx)
    color_warp = np.zeros_like(img)
    # Recast the x and y points into usable format for cv2.fillPoly()
    ploty = np.linspace(0, img.shape[0]-1, img.shape[0] )
#     print(len(Left_Lines))
    left_fitx = Left_Lines[-1].best_fit[0] *ploty*ploty + Left_Lines[-1].best_fit[1]*ploty +Left_Lines[-1].best_fit[2]
    right_fitx = Right_Lines[-1].best_fit[0] *ploty*ploty + Right_Lines[-1].best_fit[1]*ploty +Right_Lines[-1].best_fit[2]

    y_eval = np.max(ploty)*ym_per_pix
    left_fit_cr = np.polyfit(ploty*ym_per_pix, left_fitx*xm_per_pix, 2)
    left_curvature = (1+(2*left_fit_cr[0]*y_eval+left_fit_cr[1])**2)**1.5/(2*np.absolute(left_fit_cr[0])) 
    right_fit_cr = np.polyfit(ploty*ym_per_pix, right_fitx*xm_per_pix, 2)
    right_curvature = (1+(2*right_fit_cr[0]*y_eval+right_fit_cr[1])**2)**1.5/(2*np.absolute(right_fit_cr[0])) 
    
    curvature = (left_curvature +right_curvature )/2/1000.0    
    car_offset = (0.5*img.shape[1] - (np.average(left_fitx)+ np.average(right_fitx))/2) *xm_per_pix
    
    pts_left = np.array([np.transpose(np.vstack([left_fitx, ploty]))])
    pts_right = np.array([np.flipud(np.transpose(np.vstack([right_fitx,ploty])))])
    pts = np.hstack((pts_left, pts_right))
    
    str1 = " Radius of Curvature = %.3f" %curvature + 'km' 
    if car_offset> 0:
        str2 = " Vehicle is %.2f" %car_offset + 'm right of lane center'
    else:
        str2 = " Vehicle is %.2f" %-car_offset + 'm left of lane center'
    # Draw the lane onto the warped blank image
    cv2.fillPoly(color_warp, np.int_([pts]), (0,255, 0))

    # Warp the blank back to original image space using inverse perspective matrix (Minv)
    newwarp = cv2.warpPerspective(color_warp, Minv, (img.shape[1], img.shape[0])) 
    
    result = cv2.addWeighted(undistorted, 1, newwarp, 0.3, 0)
    font = cv2.FONT_HERSHEY_SIMPLEX
    cv2.putText(result,str1,(10,60), font, 2, (255, 255, 255), 4, cv2.LINE_AA)
    cv2.putText(result,str2,(10,130), font, 2, (255, 255, 255), 4, cv2.LINE_AA)
    
    return result
```

## Video Frame Process Pipeline

### Plotting sliding windows, same function rewrite in search_new_poly without visualization

```python
### Plotting sliding windows, same function rewrite in search_new_poly without visualization
img = cv2.imread('test_images/test1.jpg')
combined_binary, sxbinary, sybinary, h_binary, l_binary, s_binary, ss_binary, vv_binary = binary_threshold(img, mtx, dist)
binary_warped, M, Minv = perspective_transform(combined_binary,src_pts,dst_pts)


def slidewindow_find_lane_pixels(binary_warped):
    # Take a histogram of the bottom half of the image
    histogram = np.sum(binary_warped[binary_warped.shape[0]//2:,:], axis=0)
    # Create an output image to draw on and visualize the result
    out_img = np.dstack((binary_warped, binary_warped, binary_warped))
    # Find the peak of the left and right halves of the histogram
    # These will be the starting point for the left and right lines
    midpoint = np.int(histogram.shape[0]//2)
    leftx_base = np.argmax(histogram[:midpoint])
    rightx_base = np.argmax(histogram[midpoint:]) + midpoint

    # HYPERPARAMETERS
    # Choose the number of sliding windows
    nwindows = 9
    # Set the width of the windows +/- margin
    margin = 100
    # Set minimum number of pixels found to recenter window
    minpix = 50

    # Set height of windows - based on nwindows above and image shape
    window_height = np.int(binary_warped.shape[0]//nwindows)
    # Identify the x and y positions of all nonzero pixels in the image
    nonzero = binary_warped.nonzero()
    nonzeroy = np.array(nonzero[0])
    nonzerox = np.array(nonzero[1])
    # Current positions to be updated later for each window in nwindows
    leftx_current = leftx_base
    rightx_current = rightx_base

    # Create empty lists to receive left and right lane pixel indices
    left_lane_inds = []
    right_lane_inds = []

    # Step through the windows one by one
    for window in range(nwindows):
        # Identify window boundaries in x and y (and right and left)
        win_y_low = binary_warped.shape[0] - (window+1)*window_height
        win_y_high = binary_warped.shape[0] - window*window_height
        ### TO-DO: Find the four below boundaries of the window ###
        win_xleft_low = leftx_current - margin  # Update this
        win_xleft_high = leftx_current + margin  # Update this
        win_xright_low = rightx_current -margin  # Update this
        win_xright_high = rightx_current +margin # Update this
        
        # Draw the windows on the visualization image
        cv2.rectangle(out_img,(win_xleft_low,win_y_low),
        (win_xleft_high,win_y_high),(0,255,0), 2) 
        cv2.rectangle(out_img,(win_xright_low,win_y_low),
        (win_xright_high,win_y_high),(0,255,0), 2) 
        
        ### TO-DO: Identify the nonzero pixels in x and y within the window ###
        good_left_inds = ((nonzeroy >= win_y_low) & (nonzeroy < win_y_high) & 
        (nonzerox >= win_xleft_low) &  (nonzerox < win_xleft_high)).nonzero()[0]
        good_right_inds = ((nonzeroy >= win_y_low) & (nonzeroy < win_y_high) & 
        (nonzerox >= win_xright_low) &  (nonzerox < win_xright_high)).nonzero()[0]
        #print(type(good_right_inds),type(good_right_inds[0]))
        # Append these indices to the lists
        left_lane_inds.append(good_left_inds)
        right_lane_inds.append(good_right_inds)
        
        ### TO-DO: If you found > minpix pixels, recenter next window ###
        ### (`right` or `leftx_current`) on their mean position ###
        if len(good_left_inds) > minpix:
            leftx_current = np.int(np.mean(nonzerox[good_left_inds]))
        if len(good_right_inds) > minpix:        
            rightx_current = np.int(np.mean(nonzerox[good_right_inds]))
        #print(len(left_lane_inds),np.concatenate(left_lane_inds).shape)
    # Concatenate the arrays of indices (previously was a list of lists of pixels)
    try:
        left_lane_inds = np.concatenate(left_lane_inds)
        right_lane_inds = np.concatenate(right_lane_inds)
    except ValueError:
        # Avoids an error if the above is not implemented fully
        pass

    # Extract left and right line pixel positions
    leftx = nonzerox[left_lane_inds]
    lefty = nonzeroy[left_lane_inds] 
    rightx = nonzerox[right_lane_inds]
    righty = nonzeroy[right_lane_inds]

    return leftx, lefty, rightx, righty, out_img


def slidewindow_fit_polynomial(binary_warped):
    # Find our lane pixels first
    leftx, lefty, rightx, righty, out_img = slidewindow_find_lane_pixels(binary_warped)

    ### TO-DO: Fit a second order polynomial to each using `np.polyfit` ###
    left_fit = np.polyfit(lefty,leftx,2)
    right_fit = np.polyfit(righty,rightx,2)

    # Generate x and y values for plotting
    ploty = np.linspace(0, binary_warped.shape[0]-1, binary_warped.shape[0] )
    try:
        left_fitx = left_fit[0]*ploty**2 + left_fit[1]*ploty + left_fit[2]
        right_fitx = right_fit[0]*ploty**2 + right_fit[1]*ploty + right_fit[2]
    except TypeError:
        # Avoids an error if `left` and `right_fit` are still none or incorrect
        print('The function failed to fit a line!')
        left_fitx = 1*ploty**2 + 1*ploty
        right_fitx = 1*ploty**2 + 1*ploty

    ## Visualization ##
    # Colors in the left and right lane regions
    out_img[lefty, leftx] = [255, 0, 0]
    out_img[righty, rightx] = [0, 0, 255]

    # Plots the left and right polynomials on the lane lines
    plt.plot(left_fitx, ploty, color='yellow')
    plt.plot(right_fitx, ploty, color='yellow')

    return out_img


out_img = slidewindow_fit_polynomial(binary_warped)

plt.imshow(out_img)
```
